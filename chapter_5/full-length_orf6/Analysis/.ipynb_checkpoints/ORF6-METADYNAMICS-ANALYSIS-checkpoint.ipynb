## This notebook is for the characterisation of full-length ORF6 

## Introduction 

This jupyter notebook contains sample code to analyse metadynamic simulations of the 61-residue ORF6 from SARS-CoV-2 in the `AMBER03ws` force field at 310 K. These simulations were performed using `GROMACS 2021.2` (1) patched with the `PLUMED` library `version 2.7.1` (2).

Amino acid sequence: `MFHLVDFQVTIAEILLIIMRTFKVSIWNLDYIINLIIKNLSKSLTENKYSQLDEEQPMEID`

## Preprocessing of the trajectories using GROMACS 

Partial trajectories for each replica were first concatenated using the `GROMACS` `gmx trjcat` command to obtain full trajectories per replica. 

Time-ordered trajectories from each replica were concatenated together to a single, large trajectory using the `GROMACS gmx trjcat` command. The `PLUMED driver` command was then used to recalcuate the weights on this large trajectory.

The weights of frames in which the protein interacts with its periodic image have already been set to zero. These frames were identified using the `GROMACS gmx mindist` command. The cut-off for self-interaction was set to 1.2 nm for `AMBER03ws`. 

From the `AMBER03ws`trajectories, less than 1.2% of the ensemble (by metadynamics weight) were removed. 

I discard a total of 35% of the trajectory from the `AMBER03ws` trajectory for equilibration (see `Blocking analysis and equilibration of weights`)
        
Parts of the analysis code have been adapted from ref (3), GitHub: https://github.com/gthh2/amyloid-beta_small_mol/blob/master/Metadynamic_metainference/Analysis/Metadynamic_metainference_analysis.ipynb

```python
#Define function to return the weighted average and standard deviation
def weighted_avg_and_std(values, weights):
    average = np.average(values, weights=weights)
    variance = np.average((values-average)**2, weights=weights)
    return (average, math.sqrt(variance))
```

## Blocking analysis and equilibration of weights 

I assessed convergence of the metadynamic simulations by calculating block errors as a function of block size for free energy histograms projected onto 14 CVs. I used a minimum block size of 6 and converted block size from frames into time according to the length of 1 replica in each force field. Block analysis scripts were taken from ref (4) and the following GitHub directory: https://github.com/fpesceKU/BLOCKING/tree/main

Length of each replica in microseconds is as follows: a03ws (0.85)

To quantiatively assess convergence I also defined a bi-exponential decay function to the free energy surface error blocking for each CV. I plot the value of scaling factor at infinite time, which shows how close our simulation is to convergence. 

```python
# Define bi-exponential decay function to fit the free energy surface blocking for each CV
def conv_fit(t, A, B, C, k1, k2):
    return A * (1 - B*np.exp(-k1 * t) - C*np.exp(-k2 * t)) # 
```

```python
# Blocking analysis and fit for a03ws orf6 

# Initial guess 
p0 = (0.7,0.1,0.1,0.01,0.001) 

# set the bounds to speed up the calculation 
bounds = ([0.3,0.1,0.1,0.01,0.001],[np.inf,1e9,1e9,1e9,1e9])

block_fes_final_a03ws_orf6 = {} # store block fes results for each CV
fit_results_dict_a03ws_orf6 = {} # store fit results for each CV 

for cv_name_a03ws_orf6, cvs_a03ws_orf6 in a03ws_orf6_cvs_dict.items():
    
    # Perform block fes calculation
    block_fes_a03ws_orf6 = BlockAnalysis(cvs_a03ws_orf6, T=310, bias=bias_a03ws_orf6, multi=128)
    block_fes_final_a03ws_orf6[cv_name_a03ws_orf6] = block_fes_a03ws_orf6
    
    # Perform curve fitting
    popt_a03ws_orf6, pcov_a03ws_orf6 = curve_fit(conv_fit, block_fes_a03ws_orf6.stat[..., 0], block_fes_a03ws_orf6.stat[..., 1], p0=p0, bounds=bounds)

    # Extracting fitted parameters
    A_fit_a03ws_orf6, B_fit_a03ws_orf6, C_fit_a03ws_orf6, k1_fit_a03ws_orf6, k2_fit_a03ws_orf6 = popt_a03ws_orf6
    
    # Store fitted parameters in the dictionary
    fit_results_dict_a03ws_orf6[cv_name_a03ws_orf6] = {
        'A_fit_a03ws_orf6': A_fit_a03ws_orf6,
        'B_fit_a03ws_orf6': B_fit_a03ws_orf6,
        'C_fit_a03ws_orf6': C_fit_a03ws_orf6,
        'k1_fit_a03ws_orf6': k1_fit_a03ws_orf6,
        'k2_fit_a03ws_orf6': k2_fit_a03ws_orf6
    }
```

```python
## Plot a03ws orf6 blocking analysis 

fig, axs = plt.subplots(nrows=5, ncols=3, dpi=300, sharex=True,figsize=(140/25.4,100/25.4))

axs = axs.ravel()
xtick_labels = [0, 0.45, 0.9]

for i, (cv_name_a03ws_orf6, block_fes_a03ws_orf6) in enumerate(block_fes_final_a03ws_orf6.items()):
    ax = axs[i]
    ax.errorbar(block_fes_a03ws_orf6.stat[...,0]/889, block_fes_a03ws_orf6.stat[...,1], block_fes_a03ws_orf6.stat[...,2], fmt='', color=a_a03ws, ecolor=a_a03ws, capsize=1,linewidth=1)
    block_fes_a03ws_orf6.SEM()

    # Uncomment to see CV and axes titles for each plot 
    
    clean_cv_name_a03ws_orf6 = cv_name_a03ws_orf6.replace('_a03ws_orf6', '')
    ax.set_title(clean_cv_name_a03ws_orf6)  
    #fig.text(0.5, 0.04, r"Block size ($\mu$s)", ha="center")
    #fig.text(0.04, 0.5, r"$\langle$err(FES)$\rangle$ (kJ mol$^{-1}$)", va="center", rotation="vertical")

    # Add fitted curve
    popt_a03ws_orf6 = [fit_results_dict_a03ws_orf6[cv_name_a03ws_orf6][param_a03ws_orf6] for param_a03ws_orf6 in ['A_fit_a03ws_orf6', 'B_fit_a03ws_orf6', 'C_fit_a03ws_orf6', 'k1_fit_a03ws_orf6', 'k2_fit_a03ws_orf6']]
    fitted_curve_a03ws_orf6 = conv_fit(block_fes_a03ws_orf6.stat[..., 0], *popt_a03ws_orf6)
    ax.plot(block_fes_a03ws_orf6.stat[..., 0]/889, fitted_curve_a03ws_orf6, color='black', linewidth=1)
    # Add the horizontal dashed line
    ax.plot(block_fes_a03ws_orf6.stat[..., 0]/889, np.ones_like(block_fes_a03ws_orf6.stat[...,0])*fit_results_dict_a03ws_orf6[cv_name_a03ws_orf6]['A_fit_a03ws_orf6'], color='black', linestyle='--',alpha=0.4, linewidth=1)
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))
    ax.set_xticks(xtick_labels)
    if i < 6:
        ax.set_xticklabels([])
        
plt.tight_layout(rect=[0.05, 0.05, 1, 0.95]) 
```

## Assess cis-P57 sampling in each ensemble  

I defined omega as -60 < omega < 60 

Based on plot below it shows that no other angles are being sampled outside this range until you reach the range for trans.

I also split the trajectories into two halves to assess convergence. 

The proline residue is 57, so residue 56 is involved in cis/trans omega angle.

```python
# Uncertainty in proline cis % for a03ws 

for i in range(0, ome_a03ws_orf6_deg.shape[1]):
    
    a03ws_orf6_ome_reps = np.asarray(np.split(ome_a03ws_orf6_deg[:,i], n_reps))
    a03ws_orf6_ome_0 = a03ws_orf6_ome_reps.copy()[:, start_a03ws_orf6:mid_a03ws_orf6]
    a03ws_orf6_ome_1 = a03ws_orf6_ome_reps.copy()[:, mid_a03ws_orf6:end_a03ws_orf6]

    # First and second halves 
    a03ws_orf6_ome_0_w = np.zeros_like(a03ws_orf6_ome_0.flatten())
    a03ws_orf6_ome_0_w[np.where(np.logical_and(a03ws_orf6_ome_0.flatten()>=-60, a03ws_orf6_ome_0.flatten()<=60))] = 1
    a03ws_orf6_ome_1_w = np.zeros_like(a03ws_orf6_ome_1.flatten())
    a03ws_orf6_ome_1_w[np.where(np.logical_and(a03ws_orf6_ome_1.flatten()>=-60, a03ws_orf6_ome_1.flatten()<=60))] = 1

    # find where cis values are for whole trajectory 
    a03ws_orf6_ome_all_w = np.zeros_like(ome_a03ws_orf6_deg[:,i])
    a03ws_orf6_ome_all_w[np.where(np.logical_and(ome_a03ws_orf6_deg[:,i]>=-60, ome_a03ws_orf6_deg[:,i]<=60))] = 1

    # Calculate percentages
    perc_0 = (a03ws_orf6_ome_0_w*w_a03ws_orf6_eq_0.flatten()).sum()/w_a03ws_orf6_eq_0.flatten().sum()
    perc_1 = (a03ws_orf6_ome_1_w*w_a03ws_orf6_eq_1.flatten()).sum()/w_a03ws_orf6_eq_1.flatten().sum()
    perc_all = (a03ws_orf6_ome_all_w*w_a03ws_orf6_eq).sum()/w_a03ws_orf6_eq.sum()

    err_perc = weighted_avg_and_std([perc_0, perc_1], [sum_w_a03ws_orf6_eq_0, sum_w_a03ws_orf6_eq_1])[1]

    print('Res number {} is'.format(i+1), round(perc_all * 100, 2), '+/-', round(err_perc * 100, 2), '% cis')
```

## C-alpha minimum distance contact maps 

Selecting top 25,000 frames for the simulation ensembles - based on kish being 24,000

Here, I am taking a weighted ensemble, where everything effectively sums to 1, so normalising is not required. 

Contact map code adapted from GitHub: https://github.com/paulrobustelli/CHEM101.6/blob/main/CHEM101.6_Project4_Student.ipynb

```python
# a03ws contact map 
contact_maps_a03ws_orf6 = []
contact_distances_a03ws_orf6 = []

for i in range(0, 61): # 0 to 61 residues
    contact_map_a03ws_orf6 = []
    contact_distance_a03ws_orf6 = []
    for j in range(0, 61):
        if i == j:
            contacts_a03ws_orf6 = 0 # set to 0 if true since a residue can't be in contact with itself 
        else:
            dist_a03ws_orf6 = md.compute_contacts(t_top_a03ws_orf6, [[i, j]], scheme='ca') # based on c-alpha atoms 
            array_a03ws_orf6 = np.asarray(dist_a03ws_orf6[0]).astype(float)
            distance_a03ws_orf6 = np.average(array_a03ws_orf6)
            contact_distance_a03ws_orf6.append(distance_a03ws_orf6)
            contact_a03ws_orf6 = np.where(array_a03ws_orf6 < 0.5, 1, 0) # cutoff is 0.5 nm, contact =1, no contact=0 
            contacts_a03ws_orf6 = np.average(contact_a03ws_orf6) # calculate the average contact across the frames 
        contact_map_a03ws_orf6.append(contacts_a03ws_orf6)
    contact_maps_a03ws_orf6.append(contact_map_a03ws_orf6)
    contact_distances_a03ws_orf6.append(contact_distance_a03ws_orf6)
final_map_a03ws_orf6 = np.asarray(contact_maps_a03ws_orf6).astype(float)
final_distance_a03ws_orf6 = np.asarray(contact_distances_a03ws_orf6).astype(float)
```

```python
LJ_a03ws_cut = LJ_a03ws[top_frames_a03ws_orf6][:, 1:1892]
Coul_a03ws_cut = Coul_a03ws[top_frames_a03ws_orf6][:, 1:1892]
```

To convert these interaction energies into a contact map, interactions which were less than one standard deviation from the mean were assigned a value of -1, those within one standard deviation of the mean were assigned a value of 0, and those greater than one standard deviation of the mean were assigned a value of 1.

```python
### PROTEIN 

#Convert LJ_a03ws_cut into 1, 0, -1
mean = LJ_a03ws_cut.mean()
std  = LJ_a03ws_cut.std()
LJ_a03ws_cut_ones = np.zeros_like(LJ_a03ws_cut)
LJ_a03ws_cut_ones[LJ_a03ws_cut > (mean + std)] = 1
LJ_a03ws_cut_ones[LJ_a03ws_cut < (mean - std)] = -1

#Triangulate array for LJ_a03ws:
tri_LJ_a03ws_cut_ones = np.empty((len(LJ_a03ws_cut), 61, 61))
for i in range(0, len(LJ_a03ws_cut_ones)):
    data = LJ_a03ws_cut_ones[i]
    out = np.zeros((61, 61))
    inds = np.triu_indices(len(out))
    out[inds] = data
    tri_LJ_a03ws_cut_ones[i] = out

#Convert Coul_a03ws_cut into 1, 0, -1
mean = Coul_a03ws_cut.mean()
std  = Coul_a03ws_cut.std()
Coul_a03ws_cut_ones = np.zeros_like(Coul_a03ws_cut)
Coul_a03ws_cut_ones[Coul_a03ws_cut > (mean + std)] = 1
Coul_a03ws_cut_ones[Coul_a03ws_cut < (mean - std)] = -1
            
#Triangulate array for Coul_a03ws:
tri_Coul_a03ws_cut_ones = np.empty((len(Coul_a03ws_cut), 61, 61))
for i in range(0, len(Coul_a03ws_cut_ones)):
    data = Coul_a03ws_cut_ones[i]
    out = np.zeros((61, 61))
    inds = np.triu_indices(len(out))
    out[inds] = data
    tri_Coul_a03ws_cut_ones[i] = out
```

```python
#Plot a03ws alone contacts based on LJ and Coul contacts
#Note the residues in the plot are indexed from zero
#The code below puts the Coul contacts into the bottom left half of the plot
invert_a03ws = tri_Coul_a03ws_cut_ones.mean(axis=0)
tri_Coul_a03ws_cut_ones_bottom = invert_a03ws.T

#Combine both LJ and Coul into a single plot           
combo_a03ws = tri_LJ_a03ws_cut_ones.mean(axis=0)+tri_Coul_a03ws_cut_ones_bottom
np.fill_diagonal(combo_a03ws, -1)
ax = sns.heatmap(combo_a03ws, vmin=-0.4, vmax=0.4, cmap=cmap_con_a03ws)

ax.set_xticks(np.arange(0, 61, 10))
ax.set_yticks(np.arange(0, 61, 10))
#ax.set_xticklabels((0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60), rotation='horizontal')
#ax.set_yticklabels((0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60))
ax.set_xticklabels((1, 10, 20, 30, 40, 50, 60), rotation='horizontal')
ax.set_yticklabels((1, 10, 20, 30, 40, 50, 60))

ax.set_xlabel("Residue number")
ax.set_ylabel("Residue number")

# Add label to colorbar
cbar = ax.collections[0].colorbar
cbar.set_label('Interaction score')

fig = plt.gcf()
plt.gcf().set_size_inches(55/25.4, 55/25.4)
```
