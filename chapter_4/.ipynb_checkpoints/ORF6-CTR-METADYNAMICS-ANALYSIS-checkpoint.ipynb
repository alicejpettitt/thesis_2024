## ORF6-CTR metadynamics simulations analysis 

## Introduction 

This jupyter notebook contains sample code to analyse metadynamic simulations of the 21-residue N-acetylated C-terminal region (CTR) of ORF6 from SARS-CoV-2 in the `AMBER03ws`, `AMBER99SB-disp` and the `CHARMM36m` force fields at 310 K. These simulations were performed using `GROMACS 2021.2` (1) patched with the `PLUMED` library `version 2.7.1` (2).

Amino acid sequence: `Ac-SKSLTENKYSQLDEEQPMEID`

All input files to reproduce these simulations are located in both '../PLUMED_input_files/<ff>' or PLUMED NEST

## Preprocessing of the trajectories using GROMACS 

Partial trajectories for each replica were first concatenated using the `GROMACS` `gmx trjcat` command to obtain full trajectories per replica. 

Time-ordered trajectories from each replica were concatenated together to a single, large trajectory using the `GROMACS gmx trjcat` command. The `PLUMED driver` command was then used to recalcuate the weights on this large trajectory.

The weights of frames in which the peptide interacts with its periodic image have already been set to zero. These frames were identified using the `GROMACS gmx mindist` command. The cut-off for self-interaction was set to 1.2 nm for `AMBER03ws`, `AMBER99SB-disp`and 0.95 nm for `CHARMM36m`.

From the `AMBER03ws`trajectories, less than 0.002% of the ensemble (by metadynamics weight) were removed from runs 1 and 2, from the `AMBER99SB-disp`trajectories, less than 0.001% were removed and for the `CHARMM36m` trajectories, 0% were removed. 

We discard a total of 35% of the trajectory from the `AMBER03ws` and `AMBER99SB-disp` trajectories, and 25% from the `CHARMM36m` trajectories for equilibration (see `Blocking analysis and equilibration of weights`)
    
The two `AMBER03ws`trajectories are defined as run 1 and run 2, where run 2 is the simulation starting from 0% cis-P57 conformations.
    
Parts of the analysis code have been adapted from ref (3), GitHub: https://github.com/gthh2/amyloid-beta_small_mol/blob/master/Metadynamic_metainference/Analysis/Metadynamic_metainference_analysis.ipynb

```python
#Define function to return the weighted average and standard deviation
def weighted_avg_and_std(values, weights):
    average = np.average(values, weights=weights)
    variance = np.average((values-average)**2, weights=weights)
    return (average, math.sqrt(variance))
```

## Blocking analysis and equilibration of weights 

We assessed convergence of the metadynamic simulations by calculating block errors as a function of block size for free energy histograms projected onto 9 CVs. We used a minimum block size of 10 and converted block size from frames into time according to the length of 1 replica in each force field. Block analysis scripts were taken from ref (4) and the following GitHub directory: https://github.com/fpesceKU/BLOCKING/tree/main

Length of each replica in microseconds is as follows: a03ws (1.5), a99sb-disp (1.4), C36m (1.7).

```python
# normalise the weights 
w_a03ws_run1 /= w_a03ws_run1.sum() 
w_a03ws_run2 /= w_a03ws_run2.sum() 
w_a99sb /= w_a99sb.sum() 
w_c36m /= w_c36m.sum() 
```

```python
# Calculate the kish for each system (how many structures describe the ensemble) 

kish_a03ws_run1 = ((w_a03ws_run1.sum())*(w_a03ws_run1.sum()))/((w_a03ws_run1*w_a03ws_run1).sum())
print('a03ws run1 kish:', kish_a03ws_run1)

kish_a03ws_run2 = ((w_a03ws_run2.sum())*(w_a03ws_run2.sum()))/((w_a03ws_run2*w_a03ws_run2).sum())
print('a03ws run2 kish:', kish_a03ws_run2)

# Calculate kish_a99sb
kish_a99sb = ((w_a99sb.sum())*(w_a99sb.sum()))/((w_a99sb*w_a99sb).sum())
print('a99sb kish:', kish_a99sb)

# Calculate kish_c36m
kish_c36m = ((w_c36m.sum())*(w_c36m.sum()))/((w_c36m*w_c36m).sum())
print('c36m kish:', kish_c36m)
```

```python
# Define number of replicas 
n_reps = 128
```

```python
#Given that the simulations is arranged as 128 time-ordered replicas
# we split up the weights into 128 separate arrays

## a03ws run1 
w_a03ws_run1_split = np.split(w_a03ws_run1, n_reps)
# length of each replica 
length_w_a03ws_run1=(len(w_a03ws_run1)/n_reps) 
# Removing first 35% 
keep_w_a03ws_run1 = np.floor((length_w_a03ws_run1)*0.65) # Removing first 35% 
keep_w_a03ws_run1
for i in range(0, n_reps):
    #Set first values in ORF6-CTR ensemble to zero 
    w_a03ws_run1_split[i][0:(len(w_a03ws_run1)/n_reps-keep_w_a03ws_run1).astype(int)] = 0 
#Concatenate the separate arrays again
w_a03ws_run1_eq = np.hstack(w_a03ws_run1_split) 
#Renormalise metadynamics weights 
w_a03ws_run1_eq /= w_a03ws_run1_eq.sum() 
# print to check weights are normalised 
print('check a03ws_run1 weights:', sum(w_a03ws_run1_eq))

## a03ws run2 
w_a03ws_run2_split = np.split(w_a03ws_run2, n_reps)
length_w_a03ws_run2=(len(w_a03ws_run2)/n_reps) 
keep_w_a03ws_run2 = np.floor((length_w_a03ws_run2)*0.65) # Removing first 35% 
keep_w_a03ws_run2
for i in range(0, n_reps):
    w_a03ws_run2_split[i][0:(len(w_a03ws_run2)/n_reps-keep_w_a03ws_run2).astype(int)] = 0 
w_a03ws_run2_eq = np.hstack(w_a03ws_run2_split) 
w_a03ws_run2_eq /= w_a03ws_run2_eq.sum() 
print('check a03ws_run2 weights:', sum(w_a03ws_run2_eq))

## a99sb 
w_a99sb_split = np.split(w_a99sb, n_reps)
length_w_a99sb=(len(w_a99sb)/n_reps) 
keep_w_a99sb = np.floor((length_w_a99sb)*0.65) # Removing first 35% 
keep_w_a99sb
for i in range(0, n_reps):
    w_a99sb_split[i][0:(len(w_a99sb)/n_reps-keep_w_a99sb).astype(int)] = 0 
w_a99sb_eq = np.hstack(w_a99sb_split) 
w_a99sb_eq /= w_a99sb_eq.sum()  
print('check a99sb weights:', sum(w_a99sb_eq))

## c36m 
w_c36m_split = np.split(w_c36m, n_reps)
length_w_c36m=(len(w_c36m)/n_reps) 
keep_w_c36m = np.floor((length_w_c36m)*0.75) # Removing first 25% (less frames but run for longer, so we want the same amount of eq. time) 
keep_w_c36m
for i in range(0, n_reps):
    w_c36m_split[i][0:(len(w_c36m)/n_reps-keep_w_c36m).astype(int)] = 0 
w_c36m_eq = np.hstack(w_c36m_split) 
w_c36m_eq /= w_c36m_eq.sum()  
print('check c36m weights:', sum(w_c36m_eq))
```

```python
# Calculate the kish after equilibration 

# kish_a03ws run 1
kish_a03ws_run1_eq = ((w_a03ws_run1_eq.sum())*(w_a03ws_run1_eq.sum()))/((w_a03ws_run1_eq*w_a03ws_run1_eq).sum())
print('a03ws kish after eq:', kish_a03ws_run1_eq)

# kish a03ws run2
kish_a03ws_run2_eq = ((w_a03ws_run2_eq.sum())*(w_a03ws_run2_eq.sum()))/((w_a03ws_run2_eq*w_a03ws_run2_eq).sum())
print('a03ws_run2 kish after eq:', kish_a03ws_run2_eq)

# kish_a99sb 
kish_a99sb_eq = ((w_a99sb_eq.sum())*(w_a99sb_eq.sum()))/((w_a99sb_eq*w_a99sb_eq).sum())
print('a99sb kish after eq:', kish_a99sb_eq)

# kish_c36m 
kish_c36m_eq = ((w_c36m_eq.sum())*(w_c36m_eq.sum()))/((w_c36m_eq*w_c36m_eq).sum())
print('c36m kish after eq:', kish_c36m_eq)
```

To quantiatively assess convergence we also defined a bi-exponential decay function to the free energy surface error blocking for each CV. We plot the value of scaling factor at infinite time, which shows how close our simulation is to convergence. 

```python
# Define bi-exponential decay function to fit the free energy surface blocking for each CV
def conv_fit(t, A, B, C, k1, k2):
    return A * (1 - B*np.exp(-k1 * t) - C*np.exp(-k2 * t)) # 
```

```python
# Blocking analysis and fit for a03ws run1 

# Initial guess 
p0 = (0.7,0.1,0.1,0.01,0.001) 

# set the bounds to speed up the calculation 
bounds = ([0.3,0.1,0.1,0.01,0.001],[np.inf,1e9,1e9,1e9,1e9])

block_fes_final_a03ws_run1 = {} # store block fes results for each CV
fit_results_dict_a03ws_run1 = {} # store fit results for each CV 

for cv_name_a03ws_run1, cvs_a03ws_run1 in a03ws_run1_cvs_dict.items():
    
    # Perform block fes calculation
    block_fes_a03ws_run1 = BlockAnalysis(cvs_a03ws_run1, T=310, bias=bias_a03ws_run1, multi=128)
    block_fes_final_a03ws_run1[cv_name_a03ws_run1] = block_fes_a03ws_run1
    
    # Perform curve fitting
    popt_a03ws_run1, pcov_a03ws_run1 = curve_fit(conv_fit, block_fes_a03ws_run1.stat[..., 0], block_fes_a03ws_run1.stat[..., 1], p0=p0, bounds=bounds)

    # Extracting fitted parameters
    A_fit_a03ws_run1, B_fit_a03ws_run1, C_fit_a03ws_run1, k1_fit_a03ws_run1, k2_fit_a03ws_run1 = popt_a03ws_run1
    
    # Store fitted parameters in the dictionary
    fit_results_dict_a03ws_run1[cv_name_a03ws_run1] = {
        'A_fit_a03ws_run1': A_fit_a03ws_run1,
        'B_fit_a03ws_run1': B_fit_a03ws_run1,
        'C_fit_a03ws_run1': C_fit_a03ws_run1,
        'k1_fit_a03ws_run1': k1_fit_a03ws_run1,
        'k2_fit_a03ws_run1': k2_fit_a03ws_run1
    }
```

## SAXS reweighting 

To further refine our conformational ensembles, we used Bayesian/Maximum Entropy (BME) reweighting with SAXS data for the two independent `AMBER03ws` ensembles and the `CHARMM36m` ensemble. The `AMBER99SB-disp` ensemble was not reweighted due to the zeta improper dihedral angle not being converged. 

SAXS intensity curves were calculated for each structure in the ensemble using Pepsi-SAXS (5) with a weight > 0 (see below for two sets of arrays with and without frames that have a weight of 0)

We followed the method described in ref (6) and GitHub: https://github.com/KULL-Centre/papers/tree/main/2021/aSYN-ahmed-et-al to minimise the risk of overfitting by selecting different parameters for each structure. 

We used BME software from ref (7) and GitHub: https://github.com/KULL-Centre/BME to derive a new set of weights for each structure. Theta was set to 100 for each ensemble.  

To repeat the reweighting yourself, download the SAXS_calculations_Zenodo directory from Zenodo, which contains the SAXS intensities calculated for each frame in the a03ws run 1, a03ws run 2, and C36m metadynamic simulations, as well as the experimental SAXS data `saxs_exp_nac_orf6_ctr.dat` and follow the instructions in the following Jupyter notebook: https://github.com/KULL-Centre/papers/blob/main/2021/aSYN-ahmed-et-al/BME_analysis/BME_analysis_example.ipynb

<b> References <b> 

(5) Grudinin S, Garkavenko M, Kazennov A. 2017. Pepsi-SAXS: an adaptive method for rapid and accurate computation of small-angle X-ray scattering profiles. Acta Crystallogr Sect D, Struct Biol. 73:449–64. 

(6) Ahmed MC, Skaanning LK, Jussupow A, Newcombe EA, Kragelund BB, Camilloni C, Langkilde AE, Lindorff-Larsen K. 2021. Refinement of α-Synuclein Ensembles Against SAXS Data: Comparison of Force Fields and Methods. Front Mol Biosci. 8:654333. 

(7) Bottaro S, Bengtsen T, Lindorff-Larsen K. 2020. Integrating Molecular Simulation and Experimental Data: A Bayesian/Maximum Entropy Reweighting Approach. Methods Mol Biol. 2112:219–40.

```python
## define function to calculate residuals 

def calc_residuals(experimental_intensities, simulated_intensities):
    # Calculate the residuals
    residuals = experimental_intensities - simulated_intensities

    # Return the chi-squared value
    return residuals
```

## Assess cis-P57 sampling in each ensemble  

We defined omega as -60 < omega < 60 

Based on plot below it shows that no other angles are being sampled outside this range until you reach the range for trans.

We also split the trajectories into two halves to assess convergence. 

The proline residue is 57, so residue 56 is involved in cis/trans omega angle.

```python
# Uncertainty in proline cis % for a03ws run 1 

for i in range(0, ome_a03ws_run1_no_ACE_deg.shape[1]):
    
    a03ws_run1_ome_reps = np.asarray(np.split(ome_a03ws_run1_no_ACE_deg[:,i], n_reps))
    a03ws_run1_ome_0 = a03ws_run1_ome_reps.copy()[:, start_a03ws_run1:mid_a03ws_run1]
    a03ws_run1_ome_1 = a03ws_run1_ome_reps.copy()[:, mid_a03ws_run1:end_a03ws_run1]

    # First and second halves 
    a03ws_run1_ome_0_w = np.zeros_like(a03ws_run1_ome_0.flatten())
    a03ws_run1_ome_0_w[np.where(np.logical_and(a03ws_run1_ome_0.flatten()>=-60, a03ws_run1_ome_0.flatten()<=60))] = 1
    a03ws_run1_ome_1_w = np.zeros_like(a03ws_run1_ome_1.flatten())
    a03ws_run1_ome_1_w[np.where(np.logical_and(a03ws_run1_ome_1.flatten()>=-60, a03ws_run1_ome_1.flatten()<=60))] = 1

    # find where cis values are for whole trajectory 
    a03ws_run1_ome_all_w = np.zeros_like(ome_a03ws_run1_no_ACE_deg[:,i])
    a03ws_run1_ome_all_w[np.where(np.logical_and(ome_a03ws_run1_no_ACE_deg[:,i]>=-60, ome_a03ws_run1_no_ACE_deg[:,i]<=60))] = 1

    # Calculate percentages
    perc_0 = (a03ws_run1_ome_0_w*w_a03ws_run1_eq_0.flatten()).sum()/w_a03ws_run1_eq_0.flatten().sum()
    perc_1 = (a03ws_run1_ome_1_w*w_a03ws_run1_eq_1.flatten()).sum()/w_a03ws_run1_eq_1.flatten().sum()
    perc_all = (a03ws_run1_ome_all_w*w_a03ws_run1_eq).sum()/w_a03ws_run1_eq.sum()

    err_perc = weighted_avg_and_std([perc_0, perc_1], [sum_w_a03ws_run1_eq_0, sum_w_a03ws_run1_eq_1])[1]

    print('Res number {} is'.format(i + 41), round(perc_all * 100, 2), '+/-', round(err_perc * 100, 2), '% cis')
```

```python
# Exclude the ACE from top frames for contact maps 

t_top_a03ws_run1_no_ACE = t_top_a03ws_run1.atom_slice(t_top_a03ws_run1.top.select('not resid 0'))
t_top_a03ws_run2_no_ACE = t_top_a03ws_run2.atom_slice(t_top_a03ws_run2.top.select('not resid 0'))
t_top_a99sb_no_ACE = t_top_a99sb.atom_slice(t_top_a99sb.top.select('not resid 0'))
t_top_c36m_no_ACE = t_top_c36m.atom_slice(t_top_c36m.top.select('not resid 0'))
```

```python
# a03s run 1 contact map 
contact_maps_a03ws_run1 = []
contact_distances_a03ws_run1 = []

for i in range(0, 21): # 0 to 20 residues 
    contact_map_a03ws_run1 = []
    contact_distance_a03ws_run1 = []
    for j in range(0, 21):
        if i == j:
            contacts_a03ws_run1 = 0 # set to 0 if true since a residue can't be in contact with itself 
        else:
            dist_a03ws_run1 = md.compute_contacts(t_top_a03ws_run1_no_ACE, [[i, j]], scheme='ca') # based on c-alpha atoms 
            array_a03ws_run1 = np.asarray(dist_a03ws_run1[0]).astype(float)
            distance_a03ws_run1 = np.average(array_a03ws_run1)
            contact_distance_a03ws_run1.append(distance_a03ws_run1)
            contact_a03ws_run1 = np.where(array_a03ws_run1 < 0.5, 1, 0) # cutoff is 0.5 nm, contact =1, no contact=0 
            contacts_a03ws_run1 = np.average(contact_a03ws_run1) # calculate the average contact across the frames 
        contact_map_a03ws_run1.append(contacts_a03ws_run1)
    contact_maps_a03ws_run1.append(contact_map_a03ws_run1)
    contact_distances_a03ws_run1.append(contact_distance_a03ws_run1)
final_map_a03ws_run1 = np.asarray(contact_maps_a03ws_run1).astype(float)
final_distance_a03ws_run1 = np.asarray(contact_distances_a03ws_run1).astype(float)
```

## Define cis-P57 and trans-P57 weights for each forcefield ensemble 

To analyse the cis-P57 and trans-P57 conformations, we saved the weights as two separate arrays, where either the trans or cis frames were set to 0.

E.g. 

Where value is 0 in a03ws_run1_ome_all_w_P57 (means trans), set the weight to 0 in w_a03ws_run1_eq_cis, which gives an array of weights for only the cis frames 

Where value is 1 in a03ws_run1_ome_all_w_P57 (means cis), set the weight to 0 in w_a03ws_run1_eq_trans, which gives an array of weights for only the trans frames

```python
## a03ws run 1 cis weights 
w_a03ws_run1_eq_cis = np.copy(w_a03ws_run1_eq)  

# Set the values in w_a03ws_run1_eq_cis to 0 where a03ws_run1_ome_all_w_P57 is 0
w_a03ws_run1_eq_cis[a03ws_run1_ome_all_w_P57 == 0] = 0

# normalise the cis weights 
w_a03ws_run1_eq_cis /= w_a03ws_run1_eq_cis.sum()   
print('check a03ws_run1 cis weights:', sum(w_a03ws_run1_eq_cis))

#Reshape weights array to account for individual replicas
w_a03ws_run1_eq_reps_cis = np.reshape(w_a03ws_run1_eq_cis.copy(), (n_reps, int(len(w_a03ws_run1_eq_cis)/n_reps), 1))

#Make separate weight arrays for the first and second halves of the trajectory
w_a03ws_run1_eq_0_cis = w_a03ws_run1_eq_reps_cis.copy()[:, start_a03ws_run1:mid_a03ws_run1]
w_a03ws_run1_eq_1_cis = w_a03ws_run1_eq_reps_cis.copy()[:, mid_a03ws_run1:end_a03ws_run1]

#Use this to determine effective weight of first and second halves of the trajectory
sum_w_a03ws_run1_eq_0_cis = w_a03ws_run1_eq_0_cis.sum()
sum_w_a03ws_run1_eq_1_cis = w_a03ws_run1_eq_1_cis.sum()

#Normalise weights of first and second halves of the trajectory 
w_a03ws_run1_eq_0_cis /= w_a03ws_run1_eq_0_cis.sum()      
w_a03ws_run1_eq_1_cis /= w_a03ws_run1_eq_1_cis.sum() 

# reality check cis population 

# Uncertainty in proline cis % for a03ws_run1 

for i in range(0, ome_a03ws_run1_no_ACE_deg.shape[1]):
    
    #Split into first and second halves
    a03ws_run1_ome_reps = np.asarray(np.split(ome_a03ws_run1_no_ACE_deg[:,i], n_reps))
    
    #Make separate weight arrays for the first and second halves of the trajectory
    a03ws_run1_ome_0 = a03ws_run1_ome_reps.copy()[:, start_a03ws_run1:mid_a03ws_run1]
    a03ws_run1_ome_1 = a03ws_run1_ome_reps.copy()[:, mid_a03ws_run1:end_a03ws_run1]

    # find where cis values are for first and second halves
    a03ws_run1_ome_0_w = np.zeros_like(a03ws_run1_ome_0.flatten())
    a03ws_run1_ome_0_w[np.where(np.logical_and(a03ws_run1_ome_0.flatten()>=-60, a03ws_run1_ome_0.flatten()<=60))] = 1
    a03ws_run1_ome_1_w = np.zeros_like(a03ws_run1_ome_1.flatten())
    a03ws_run1_ome_1_w[np.where(np.logical_and(a03ws_run1_ome_1.flatten()>=-60, a03ws_run1_ome_1.flatten()<=60))] = 1

    # find where cis values are for whole trajectory 
    a03ws_run1_ome_all_w = np.zeros_like(ome_a03ws_run1_no_ACE_deg[:,i])
    a03ws_run1_ome_all_w[np.where(np.logical_and(ome_a03ws_run1_no_ACE_deg[:,i]>=-60, ome_a03ws_run1_no_ACE_deg[:,i]<=60))] = 1

    #get percentages
    perc_0 = (a03ws_run1_ome_0_w*w_a03ws_run1_eq_0_cis.flatten()).sum()/w_a03ws_run1_eq_0_cis.flatten().sum()
    perc_1 = (a03ws_run1_ome_1_w*w_a03ws_run1_eq_1_cis.flatten()).sum()/w_a03ws_run1_eq_1_cis.flatten().sum()
    perc_all = (a03ws_run1_ome_all_w*w_a03ws_run1_eq_cis).sum()/w_a03ws_run1_eq_cis.sum()
    err_perc = weighted_avg_and_std([perc_0, perc_1], [sum_w_a03ws_run1_eq_0_cis, sum_w_a03ws_run1_eq_1_cis])[1]

    print('Res number {} is'.format(i+41), round(perc_all*100), '+/-', round(err_perc*100), '% cis' )
```

## P57 trans vs cis secondary structure probabilities 

Calculated using DSSP module in MDTraj (8)

<b> References <b>

```python
# calc dssp for a03ws run 1 

dssp_a03ws_run1_rw = md.compute_dssp(t_a03ws_run1, simplified=True)

# Set values to 1 for coil
coil_a03ws_run1_rw = np.zeros_like(dssp_a03ws_run1_rw, dtype=int)
# Set the elements to 1 where dssp_a03ws_run1_rw is 'C'
coil_a03ws_run1_rw[dssp_a03ws_run1_rw == 'C'] = 1

# Set values to 1 for strand
strand_a03ws_run1_rw = np.zeros_like(dssp_a03ws_run1_rw, dtype=int)
# Set the elements to 1 where dssp_a03ws_run1_rw is 'E'
strand_a03ws_run1_rw[dssp_a03ws_run1_rw == 'E'] = 1

# Set values to 1 for helix
helix_a03ws_run1_rw = np.zeros_like(dssp_a03ws_run1_rw, dtype=int)
# Set the elements to 1 where dssp_a03ws_run1_rw is 'H'
helix_a03ws_run1_rw[dssp_a03ws_run1_rw == 'H'] = 1

# ensemble weighted average for coil, strand and helix (CIS)

cis_coil_a03ws_run1_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    cis_coil_a03ws_run1_rw[i] = (coil_a03ws_run1_rw[:, i] * w_a03ws_run1_rw_cis).sum()
    
cis_strand_a03ws_run1_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    cis_strand_a03ws_run1_rw[i] = (strand_a03ws_run1_rw[:, i] * w_a03ws_run1_rw_cis).sum()
    
cis_helix_a03ws_run1_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    cis_helix_a03ws_run1_rw[i] = (helix_a03ws_run1_rw[:, i] * w_a03ws_run1_rw_cis).sum()
    
# check all add to 1 
#cis_coil_a03ws_run1_rw+cis_strand_a03ws_run1_rw+cis_helix_a03ws_run1_rw

# ensemble weighted average for coil, strand and helix (trans)

trans_coil_a03ws_run1_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    trans_coil_a03ws_run1_rw[i] = (coil_a03ws_run1_rw[:, i] * w_a03ws_run1_rw_trans).sum()
    
trans_strand_a03ws_run1_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    trans_strand_a03ws_run1_rw[i] = (strand_a03ws_run1_rw[:, i] * w_a03ws_run1_rw_trans).sum()
    
trans_helix_a03ws_run1_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    trans_helix_a03ws_run1_rw[i] = (helix_a03ws_run1_rw[:, i] * w_a03ws_run1_rw_trans).sum()
    
# check all add to 1 
#trans_coil_a03ws_run1_rw+trans_strand_a03ws_run1_rw+trans_helix_a03ws_run1_rw

# Split traj for coil, helix and strand 
coil_a03ws_run1_rw_reps = np.reshape(coil_a03ws_run1_rw, (n_reps, int(len(w_a03ws_run1_rw)/n_reps), 22))
coil_a03ws_run1_rw_0 = coil_a03ws_run1_rw_reps.copy()[:, start_a03ws_run1_rw:mid_a03ws_run1_rw]
coil_a03ws_run1_rw_1 = coil_a03ws_run1_rw_reps.copy()[:, mid_a03ws_run1_rw:end_a03ws_run1_rw]

helix_a03ws_run1_rw_reps = np.reshape(helix_a03ws_run1_rw, (n_reps, int(len(w_a03ws_run1_rw)/n_reps), 22))
helix_a03ws_run1_rw_0 = helix_a03ws_run1_rw_reps.copy()[:, start_a03ws_run1_rw:mid_a03ws_run1_rw]
helix_a03ws_run1_rw_1 = helix_a03ws_run1_rw_reps.copy()[:, mid_a03ws_run1_rw:end_a03ws_run1_rw]

strand_a03ws_run1_rw_reps = np.reshape(strand_a03ws_run1_rw, (n_reps, int(len(w_a03ws_run1_rw)/n_reps), 22))
strand_a03ws_run1_rw_0 = strand_a03ws_run1_rw_reps.copy()[:, start_a03ws_run1_rw:mid_a03ws_run1_rw]
strand_a03ws_run1_rw_1 = strand_a03ws_run1_rw_reps.copy()[:, mid_a03ws_run1_rw:end_a03ws_run1_rw]

# cis coil error 
err_coil_a03ws_run1_rw_cis = np.zeros(22)

for i in range(0,22):
    coil_a03ws_run1_rw_tuple_cis = weighted_avg_and_std([
        (coil_a03ws_run1_rw_0[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_0_cis).sum(),
        (coil_a03ws_run1_rw_1[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_1_cis).sum()
    ], [sum_w_a03ws_run1_rw_0_cis, sum_w_a03ws_run1_rw_1_cis])
    
    # Extract the relevant value from the result tuple
    coil_a03ws_run1_rw_result_cis = coil_a03ws_run1_rw_tuple_cis[1]  
    
    err_coil_a03ws_run1_rw_cis[i] = coil_a03ws_run1_rw_result_cis
    
# trans coil error 
err_coil_a03ws_run1_rw_trans = np.zeros(22)

for i in range(0,22):
    coil_a03ws_run1_rw_tuple_trans = weighted_avg_and_std([
        (coil_a03ws_run1_rw_0[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_0_trans).sum(),
        (coil_a03ws_run1_rw_1[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_1_trans).sum()
    ], [sum_w_a03ws_run1_rw_0_trans, sum_w_a03ws_run1_rw_1_trans])
    
    # Extract the relevant value from the result tuple
    coil_a03ws_run1_rw_result_trans = coil_a03ws_run1_rw_tuple_trans[1]  
    
    err_coil_a03ws_run1_rw_trans[i] = coil_a03ws_run1_rw_result_trans
    
# cis strand error 
err_strand_a03ws_run1_rw_cis = np.zeros(22)

for i in range(0,22):
    strand_a03ws_run1_rw_tuple_cis = weighted_avg_and_std([
        (strand_a03ws_run1_rw_0[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_0_cis).sum(),
        (strand_a03ws_run1_rw_1[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_1_cis).sum()
    ], [sum_w_a03ws_run1_rw_0_cis, sum_w_a03ws_run1_rw_1_cis])
    
    # Extract the relevant value from the result tuple
    strand_a03ws_run1_rw_result_cis = strand_a03ws_run1_rw_tuple_cis[1]  
    
    err_strand_a03ws_run1_rw_cis[i] = strand_a03ws_run1_rw_result_cis
    
# trans strand error 
err_strand_a03ws_run1_rw_trans = np.zeros(22)

for i in range(0,22):
    strand_a03ws_run1_rw_tuple_trans = weighted_avg_and_std([
        (strand_a03ws_run1_rw_0[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_0_trans).sum(),
        (strand_a03ws_run1_rw_1[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_1_trans).sum()
    ], [sum_w_a03ws_run1_rw_0_trans, sum_w_a03ws_run1_rw_1_trans])
    
    # Extract the relevant value from the result tuple
    strand_a03ws_run1_rw_result_trans = strand_a03ws_run1_rw_tuple_trans[1] 
    
    err_strand_a03ws_run1_rw_trans[i] = strand_a03ws_run1_rw_result_trans
    
# cis helix error 
err_helix_a03ws_run1_rw_cis = np.zeros(22)

for i in range(0,22):
    helix_a03ws_run1_rw_tuple_cis = weighted_avg_and_std([
        (helix_a03ws_run1_rw_0[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_0_cis).sum(),
        (helix_a03ws_run1_rw_1[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_1_cis).sum()
    ], [sum_w_a03ws_run1_rw_0_cis, sum_w_a03ws_run1_rw_1_cis])
    
    # Extract the relevant value from the result tuple
    helix_a03ws_run1_rw_result_cis = helix_a03ws_run1_rw_tuple_cis[1] 
    
    err_helix_a03ws_run1_rw_cis[i] = helix_a03ws_run1_rw_result_cis
    
# trans helix error 
err_helix_a03ws_run1_rw_trans = np.zeros(22)

for i in range(0,22):
    helix_a03ws_run1_rw_tuple_trans = weighted_avg_and_std([
        (helix_a03ws_run1_rw_0[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_0_trans).sum(),
        (helix_a03ws_run1_rw_1[:, :, i][:, :, np.newaxis] * w_a03ws_run1_rw_1_trans).sum()
    ], [sum_w_a03ws_run1_rw_0_trans, sum_w_a03ws_run1_rw_1_trans])
    
    # Extract the relevant value from the result tuple
    helix_a03ws_run1_rw_result_trans = helix_a03ws_run1_rw_tuple_trans[1]  
    
    err_helix_a03ws_run1_rw_trans[i] = helix_a03ws_run1_rw_result_trans
```

```python
# calc dssp for c36m 

dssp_c36m_rw = md.compute_dssp(t_c36m, simplified=True)

# Set values to 1 for coil
coil_c36m_rw = np.zeros_like(dssp_c36m_rw, dtype=int)
# Set the elements to 1 where dssp_c36m_rw is 'C'
coil_c36m_rw[dssp_c36m_rw == 'C'] = 1

# Set values to 1 for strand
strand_c36m_rw = np.zeros_like(dssp_c36m_rw, dtype=int)
# Set the elements to 1 where dssp_c36m_rw is 'E'
strand_c36m_rw[dssp_c36m_rw == 'E'] = 1

# Set values to 1 for helix
helix_c36m_rw = np.zeros_like(dssp_c36m_rw, dtype=int)
# Set the elements to 1 where dssp_c36m_rw is 'H'
helix_c36m_rw[dssp_c36m_rw == 'H'] = 1

# ensemble weighted average for coil, strand and helix (CIS)

cis_coil_c36m_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    cis_coil_c36m_rw[i] = (coil_c36m_rw[:, i] * w_c36m_rw_cis).sum()
    
cis_strand_c36m_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    cis_strand_c36m_rw[i] = (strand_c36m_rw[:, i] * w_c36m_rw_cis).sum()
    
cis_helix_c36m_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    cis_helix_c36m_rw[i] = (helix_c36m_rw[:, i] * w_c36m_rw_cis).sum()
    
# check all add to 1 
#cis_coil_c36m_rw+cis_strand_c36m_rw+cis_helix_c36m_rw

# ensemble weighted average for coil, strand and helix (trans)

trans_coil_c36m_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    trans_coil_c36m_rw[i] = (coil_c36m_rw[:, i] * w_c36m_rw_trans).sum()
    
trans_strand_c36m_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    trans_strand_c36m_rw[i] = (strand_c36m_rw[:, i] * w_c36m_rw_trans).sum()
    
trans_helix_c36m_rw = np.zeros((22), dtype=object)

for i in range(0, 22):
    trans_helix_c36m_rw[i] = (helix_c36m_rw[:, i] * w_c36m_rw_trans).sum()
    
# check all add to 1 
#trans_coil_c36m_rw+trans_strand_c36m_rw+trans_helix_c36m_rw

# Split traj for coil, helix and strand 
coil_c36m_rw_reps = np.reshape(coil_c36m_rw, (n_reps, int(len(w_c36m_rw)/n_reps), 22))
coil_c36m_rw_0 = coil_c36m_rw_reps.copy()[:, start_c36m_rw:mid_c36m_rw]
coil_c36m_rw_1 = coil_c36m_rw_reps.copy()[:, mid_c36m_rw:end_c36m_rw]

helix_c36m_rw_reps = np.reshape(helix_c36m_rw, (n_reps, int(len(w_c36m_rw)/n_reps), 22))
helix_c36m_rw_0 = helix_c36m_rw_reps.copy()[:, start_c36m_rw:mid_c36m_rw]
helix_c36m_rw_1 = helix_c36m_rw_reps.copy()[:, mid_c36m_rw:end_c36m_rw]

strand_c36m_rw_reps = np.reshape(strand_c36m_rw, (n_reps, int(len(w_c36m_rw)/n_reps), 22))
strand_c36m_rw_0 = strand_c36m_rw_reps.copy()[:, start_c36m_rw:mid_c36m_rw]
strand_c36m_rw_1 = strand_c36m_rw_reps.copy()[:, mid_c36m_rw:end_c36m_rw]

# cis coil error 
err_coil_c36m_rw_cis = np.zeros(22)

for i in range(0,22):
    coil_c36m_rw_tuple_cis = weighted_avg_and_std([
        (coil_c36m_rw_0[:, :, i][:, :, np.newaxis] * w_c36m_rw_0_cis).sum(),
        (coil_c36m_rw_1[:, :, i][:, :, np.newaxis] * w_c36m_rw_1_cis).sum()
    ], [sum_w_c36m_rw_0_cis, sum_w_c36m_rw_1_cis])
    
    # Extract the relevant value from the result tuple
    coil_c36m_rw_result_cis = coil_c36m_rw_tuple_cis[1]  
    
    err_coil_c36m_rw_cis[i] = coil_c36m_rw_result_cis
    
# trans coil error 
err_coil_c36m_rw_trans = np.zeros(22)

for i in range(0,22):
    coil_c36m_rw_tuple_trans = weighted_avg_and_std([
        (coil_c36m_rw_0[:, :, i][:, :, np.newaxis] * w_c36m_rw_0_trans).sum(),
        (coil_c36m_rw_1[:, :, i][:, :, np.newaxis] * w_c36m_rw_1_trans).sum()
    ], [sum_w_c36m_rw_0_trans, sum_w_c36m_rw_1_trans])
    
    # Extract the relevant value from the result tuple
    coil_c36m_rw_result_trans = coil_c36m_rw_tuple_trans[1] 
    
    err_coil_c36m_rw_trans[i] = coil_c36m_rw_result_trans
    
# cis strand error 
err_strand_c36m_rw_cis = np.zeros(22)

for i in range(0,22):
    strand_c36m_rw_tuple_cis = weighted_avg_and_std([
        (strand_c36m_rw_0[:, :, i][:, :, np.newaxis] * w_c36m_rw_0_cis).sum(),
        (strand_c36m_rw_1[:, :, i][:, :, np.newaxis] * w_c36m_rw_1_cis).sum()
    ], [sum_w_c36m_rw_0_cis, sum_w_c36m_rw_1_cis])
    
    # Extract the relevant value from the result tuple
    strand_c36m_rw_result_cis = strand_c36m_rw_tuple_cis[1] 
    
    err_strand_c36m_rw_cis[i] = strand_c36m_rw_result_cis
    
# trans strand error 
err_strand_c36m_rw_trans = np.zeros(22)

for i in range(0,22):
    strand_c36m_rw_tuple_trans = weighted_avg_and_std([
        (strand_c36m_rw_0[:, :, i][:, :, np.newaxis] * w_c36m_rw_0_trans).sum(),
        (strand_c36m_rw_1[:, :, i][:, :, np.newaxis] * w_c36m_rw_1_trans).sum()
    ], [sum_w_c36m_rw_0_trans, sum_w_c36m_rw_1_trans])
    
    # Extract the relevant value from the result tuple
    strand_c36m_rw_result_trans = strand_c36m_rw_tuple_trans[1] 
    err_strand_c36m_rw_trans[i] = strand_c36m_rw_result_trans
    
# cis helix error 
err_helix_c36m_rw_cis = np.zeros(22)

for i in range(0,22):
    helix_c36m_rw_tuple_cis = weighted_avg_and_std([
        (helix_c36m_rw_0[:, :, i][:, :, np.newaxis] * w_c36m_rw_0_cis).sum(),
        (helix_c36m_rw_1[:, :, i][:, :, np.newaxis] * w_c36m_rw_1_cis).sum()
    ], [sum_w_c36m_rw_0_cis, sum_w_c36m_rw_1_cis])
    
    # Extract the relevant value from the result tuple
    helix_c36m_rw_result_cis = helix_c36m_rw_tuple_cis[1]  
    
    err_helix_c36m_rw_cis[i] = helix_c36m_rw_result_cis
    
# trans helix error 
err_helix_c36m_rw_trans = np.zeros(22)

for i in range(0,22):
    helix_c36m_rw_tuple_trans = weighted_avg_and_std([
        (helix_c36m_rw_0[:, :, i][:, :, np.newaxis] * w_c36m_rw_0_trans).sum(),
        (helix_c36m_rw_1[:, :, i][:, :, np.newaxis] * w_c36m_rw_1_trans).sum()
    ], [sum_w_c36m_rw_0_trans, sum_w_c36m_rw_1_trans])
    
    # Extract the relevant value from the result tuple
    helix_c36m_rw_result_trans = helix_c36m_rw_tuple_trans[1]  
    
    err_helix_c36m_rw_trans[i] = helix_c36m_rw_result_trans
```
